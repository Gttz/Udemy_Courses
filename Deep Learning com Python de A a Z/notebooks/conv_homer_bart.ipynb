## Base de imagens homer_bart


```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.layers.normalization import BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from keras.preprocessing import image
```

```python
CNN = Sequential()
```

###### Etapa 1 - Convolution
Detector de Caractersticas e Mapa de Caractersticas

    Nmero de filtros (32)
    Dimenses do detector de caractersticas (3,3)
    Definio da altura/largura e canais RGB (64, 64, 3)
    Funo de ativao para retirar os valores negativos da imagem - 'relu'

```python
CNN.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
```

###### Etapa 2 - Max Pooling
Reduo do tamanho da imagem focando nas caracteristicas mais importantes

    Definio da matriz com o total de 4 pixels (2, 2)

```python
CNN.add(MaxPooling2D(pool_size = (2, 2)))
```

###### Camadas ocultas
    Definio das camadas ocultas

```python
CNN.add(Conv2D(32, (3, 3), activation = 'relu'))
CNN.add(MaxPooling2D(pool_size = (2, 2)))
```

###### Etapa 3 - Flattening
    
    Transformao da matriz para um vetor para entrar na camada da RN

```python
CNN.add(Flatten())
```

###### Etapa 4 - Redes Neurais Densas

Parmetros da rede neural artificial:

    Dense - Todos os neurnios conectados
    units - Quantidade de neurnios que fazem parte da camada oculta - (128)
    activation - Funo de ativao que ser inserida - 'relu'
    
Parmetros do otimizador:

    optimizer - descida do gradiente e descida do gradiente estocstica (adam  uma otimizao do estocstica)
    loss - Funo de perda (binary_crossentropy pois s h uma sada)
    metrics - Mtrica de avaliao (obs - pode ser colocado mais de um)

```python
# Camada de entrada
CNN.add(Dense(units = 4, activation = 'relu'))
# Camada oculta
CNN.add(Dense(units = 4, activation = 'relu'))
# Camada de saida (classificao binria - Homer/Bart)
CNN.add(Dense(units = 1, activation = 'sigmoid'))
# Compilao
CNN.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

###### Etapa 5 - Pr-processamento das imagens

Paramtros Do ImageDataGenerator:

    rescale - Transforma o tamanho da imagem (normalizao dos dados)
    rotation_range - grau da rotao da imagem
    horizontal_flip - Girar as imagens horizontalmente
    shear_range - Transformaes geometricas aleatrias
    height_shift_range - Faixa da mudana da altura da imagem
    zoom_range - Imagens que sero aplicadas o zoom

```python
train_datagen = ImageDataGenerator(rescale = 1./255, 
                                   rotation_range =7, 
                                   horizontal_flip = True, 
                                   shear_range = 0.2,
                                   height_shift_range = 0.07,
                                   zoom_range = 0.2)
# Tambm deve ser feito para o dataset de teste
test_datagen = ImageDataGenerator(rescale = 1./255)
```

###### Etapa 6 - Diretrio das imagens de treino e teste

Paramtros do flow_from_directory:

    target_size - tamanho das imagens (lembrando que deve ser do mesmo tamanho da camada de entrada)
    batch_size - tamanho do batch
    class_mode -modo da classe (binrio)

```python
training_set = train_datagen.flow_from_directory('training_set', 
                                                 target_size= (64, 64), 
                                                 batch_size= 10, 
                                                 class_mode='binary')
```

```python
test_set = test_datagen.flow_from_directory('test_set', 
                                            target_size= (64, 64), 
                                            batch_size= 10, 
                                            class_mode='binary')
```

###### Etapa 7 - Treino e Teste

Paramtros do fit_generator

    steps_per_epoch - nmero de imagens de treino
    epochs - nmero de pocas
    validation_data - base de dados de teste
    validation_steps - nmero de imagens para teste

```python
CNN.fit_generator(training_set, steps_per_epoch= 196, epochs = 5, validation_data= test_set, validation_steps= 73)
```

```python
training_set.class_indices
```

###### Previso com imagens individuais
    bart = 0 (nmeros abaixo de 0.50  considerado bart)
    homer = 1 (nmeros acima de 0.50  considerado homer)

Teste com Bart

```python
# Pr processamento da imagem
test_bart = image.load_img('test_set/bart/bart36.bmp',
                           target_size = (64, 64))
test_bart = image.img_to_array(test_bart)
test_bart /= 255
test_bart = np.expand_dims(test_bart, axis = 0)
```

```python
y_test = CNN.predict(test_bart)
```

```python
if y_test < 0.50:
    print(f'A imagem foi classificada com o valor de {y_test}, tendendo a ser o bart')
elif y_test >= 0.50:
    print(f'A imagem foi classificada com o valor de {y_test}, tendendo a ser o homer')
```

Teste com Homer

```python
# Pr processamento da imagem
test_homer = image.load_img('test_set/homer/homer5.bmp',
                           target_size = (64, 64))
test_homer = image.img_to_array(test_homer)
test_homer /= 255
test_homer = np.expand_dims(test_homer, axis = 0)
```

```python
y_test = CNN.predict(test_homer)
```

```python
if y_test < 0.50:
    print(f'A imagem foi classificada com o valor de {y_test}, tendendo a ser o bart')
elif y_test >= 0.50:
    print(f'A imagem foi classificada com o valor de {y_test}, tendendo a ser o homer')
```
