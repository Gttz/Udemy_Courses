### Restricted Boltzmann Machines


```python
from rbm import RBM
import numpy as np
```

```python
rbm = RBM(num_visible = 6, num_hidden = 2)
```

###### Declarao dos filmes e treinamento do algortimo

```python
df = np.array([[1,1,1,0,0,0], # Terror 3/3 e comdia 0/3
               [1,0,1,0,0,0], # Terror 2/3 e comdia 0/3
               [1,1,1,0,0,0], # Terror 3/3 e comdia 0/3
               [0,0,1,1,1,1], # Terror 1/3 e comdia 3/3
               [0,0,1,1,0,1], # Terror 1/3 e comdia 2/3
               [0,0,1,1,0,1]]) # Terror 1/3 e comdia 2/3

filmes = ["A bruxa", "Invocao do mal", "O chamado",
          "Se beber no case", "Gente grande", "American pie"]
```

```python
rbm.train(df, max_epochs = 5000)
```

```python
rbm.weights
```

Aqui podemos observar que o algoritmo conseguiu descobrir sozinho cada uma das categorias (terror e comdia), lembrando que no passamos quais eram as categorias dos filmes.

###### Criao dos usurios para os filmes de terror e comdia

```python
usuario1 = np.array([[1,1,0,1,0,0]])
usuario2 = np.array([[0,0,0,1,1,0]])
```

```python
# Foco em terror
rbm.run_visible(usuario1)
```

```python
# Foco em comdia
rbm.run_visible(usuario2)
```

Podemos observar que o algoritmo recomendou filmes de terror para o usurio 1 e filmes de comdia para o usurio 2

```python
# teste do usurio 1
camada_escondida1 = np.array([[0, 1]])
recomendacao1 = rbm.run_hidden(camada_escondida1)
recomendacao1
```

```python
# teste do usurio 2
camada_escondida2 = np.array([[1, 0]])
recomendacao2 = rbm.run_hidden(camada_escondida2)
recomendacao2
```

###### Filmes recomendados para cada usurio

```python
for i in range(len(usuario1[0])):
    # print(usuario1[0,i])
    if usuario1[0,i] == 0 and recomendacao1[0,i] == 1:
        print(filmes[i])
```

Com isso, observamos que para o usurio 1 foi recomendado o filme 'O chamado'

```python
for i in range(len(usuario2[0])):
    # print(usuario1[0,i])
    if usuario2[0,i] == 0 and recomendacao2[0,i] == 1:
        print(filmes[i])
```

Com isso, observamos que para o usurio 2 foi recomendado os filmes 'O chamado' e 'American pie'
